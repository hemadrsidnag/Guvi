```python
#defining a function Register for new registration
def register():
  db=open("datafile.txt","r")
  Username=input("Create Username: ")
  Password=input("Create Password: ")
  pass1=input("Confirm Password: ")
  d=[]
  f=[]
  dec=0
  dec1=0
  dec2=0
  for i in db:
    a,b = i.split(", ")
    b=b.strip()
    d.append(a)
    f.append(b)
  #creating a dictionary to store username and password
  data= dict(zip(d,f))

  for j in range(len(Username)-1):
    if Username[j]=='@' and Username[j+1]=='.':
      dec=1
# counters to see if the compiler ran through the condition
# assigning 1 if it met each condition of the password requirement
  sco=0
  sco1=0
  sco2=0
  sco3=0
  sco4=0

#running loop through each letter to check if the password meets required strength
  for k in range(len(Password)):
    if Password[k].isdigit():
      sco=1
      
    elif Password[k].isalpha():
      sco1=1
    
    else:
      sco4=1
      
  for l in range(len(Password)):
    if Password[l].isupper():
      sco2=1
    elif Password[l].islower():
      sco3=1
 
  if sco+sco1+sco2+sco3+sco4<5:
    print('Password is not strong enough. \n Password Must contain:\n\
    1. One Uppercase letter\n\
    2. One Lowercase letter\n\
    3. One Digit \n\
    4. One Special Character')
    register()

  if Username.index('@')>Username.index('.'):
    dec1=1

  if Username[0].isalnum():
    dec2=1

  if Password!=pass1:
    print('Password Does NOT match, please try again.')
    register()

  else:

    if Username.count('@')!=1 and Username.count('.')!=1:
      print(' @ and . not found.')

    else:
      if Username in d:
        print('Username already exists. Please try again.')
        register()

      elif len(Password)<5 or len(Password)>16:
        print('The Password must be between the lenght 5 to 16.')
        register()
      
      elif dec==1:
        print('@ cannot be next to \".\" \n Restart')
        register()

      elif dec1==1:
        print('@ should come before \".\" \n Restart')
        register()
        
      elif dec2!=1:
        print('Username should not begin with a special character. \n Restart ')
        register()

      else:
        db=open("datafile.txt","a")
        db.write(Username+ ", " + Password+ '\n')
        print('Successful!')
        

#defining function for sign-in
def access():
  
  db=open("datafile.txt","r")
  Username=input("Enter Username: ")
  Password=input("Enter Password: ")
  
  if len(Username or Password)<1:
    print('Error. try again')
    access()
  else:
    d=[]
    f=[]
    for i in db:
      a,b = i.split(", ")
      b=b.strip()
      d.append(a)
      f.append(b)
    data= dict(zip(d,f))

    try:
      if data[Username]:
        try:
          if Password==data[Username]:
            print('Login Successful')
            print('Hello, ',Username, ' Hope you are doing good')

        except:
          print("Wrong Password")
    except:
      print("Username Doesn't Exist. Try Again")
      access()

#defining funciton to retrieve forgotten password
def forgot():
  
  db=open('datafile.txt','r')
  Username= input('Please enter your Username: ')
  if len(Username or Password)<1:
   print('Error. try again')
   forgot()

  else:
    d=[]
    f=[]
    for i in db:
      a,b = i.split(", ")
      b=b.strip()
      d.append(a)
      f.append(b)
    data= dict(zip(d,f))

    #checking if the username exists in database, or else asking ot register.
    if Username in data:
      print('Your Password is: ',data[Username])
    else:
      print('Username not found.\nPlease Register')
      register()

def portal():
  print('WELCOME TO THE LOGIN PORTAL')
  inp=input('PLEASE ENTER \n1 FOR LOGIN \n2 FOR REGISTER AND \n3 FORGOT PASSWORD\n: ')

  if inp=='1':
    access()
  elif inp=='2':
    register()
  elif inp=='3':
    forgot()
  else:
    print('Choose between 1 and 3 only.\nTry Again\n')
    portal()


portal()
```
